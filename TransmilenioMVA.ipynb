```
RutasEstacionesTM = RutasEstacionesTM.groupby('Ruta',as_index=False).agg(list)
RutasEstacionesTM
```

```
RutasEstacionesTM = RutasEstacionesTM.rename(columns={'Posici√≥n': 'Position','Nombre Nodo':'ListaParadas'})
EstacionesRutasTM = EstacionesRutasTM.rename(columns={'Nombre Nodo': 'Estacion','Ruta':'Rutas'})

```

```
Rutas = pd.Series([element.lstrip("[1234567890]") for element in RutasEstacionesTM['Ruta']])
```

```
Rutas = pd.Series([element.rstrip("abcdefghijklmenoprstuvwxyzABCDEFGHIJKLMNOPQRSTUWVXYZ. ") for element in Rutas])
```

```
Rutas = pd.Series([element.lstrip(" -") for element in Rutas])
```

```
RutasEstacionesTM['Ruta'] = Rutas
```
