Se observa que no es suficiente información para poder mostrar de una manera informativa al usuario por lo que se busca nuevamente una fuente que nos brinde las estaciones como las rutas de esa estacion, y también las rutas como las paradas de esas rutas.

```
RutasEstacionesTM = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Datos/RutasEstacionesTM.csv",encoding='latin1')
```

```
RutasEstacionesTM
```

```
#Eliminación Columnas (Atributos) no necesarios.

DataSetTM = RutasEstacionesTM.drop(columns=['Tipo Servicio','Nodo','Línea']); 
DataSetTM = DataSetTM.rename(columns={'Nombre Nodo': 'Estaciones','Ruta':'Rutas'})
```

### Manipulación y Transformacion de datos
***
Este nuevo dataSet tiene los datos necesarios pero hay que realizar algunas transformaciones a estos. Eso es lo que se hará a continuación:

```
#Agrupación por Estaciones, se enlistan las rutas de cada estación.

EstacionesListRutasTM = DataSetTM.groupby('Estaciones',as_index=False).agg(list);
EstacionesListRutasTM
```

```
#Agrupación por Rutas, se muestran las paradas de cada Ruta

RutasListParadasTM = DataSetTM.groupby('Rutas',as_index=False).agg(list)
RutasListParadasTM
```

#### Tratamiento de datos.
***
Se observo que los datos no estan limpios para ser introduciodos en la aplicación (Tienen datos inecesarios como un codigo al principio o estaciones de partida). Se procedió a hacer una limpieza de datos para que unicamente aparecieran el identificador de la Ruta (Ejm: K16,B23,F14). 

Además se observo que las estaciones también estaban separadas por vagones y por sentido, esto es importante para posteriores versiones del proyecto pero por ahora solo queremos que se separen por una unica estación, se procedio a 
limpiar estos datos también.

#### Tratamiento a las Rutas
***


```
#Se hace un list Comprehension a la lista de Rutas, se eliminan los numeros y corchetes que aparecen al principio

RutasListParadasTM['Rutas'] = pd.Series([route.lstrip("[0123456789]") for route in RutasListParadasTM['Rutas']]) 
#lstrip Elimina los caracteres que se encuentran a la izquierda y que coinciden con el input.
#------------------------------------------------------------------------------------------------------------------#
#Otra forma de hacerlo.
RutasListParadasTM['Rutas'] = RutasListParadasTM['Rutas'].str.lstrip("[0123456789]")

RutasListParadasTM
```

```
#Se hace algo parecido para eliminar el punto de partida de la ruta que se encuentra al final del String. Se eliminan también espacios.
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ._-éÁ "
RutasListParadasTM['Rutas'] = pd.Series([route.rstrip(alphabet+alphabet.lower()) for route in RutasListParadasTM['Rutas']])
#lstrip Elimina los caracteres que se encuentran a la derecha y que coinciden con el input.
#------------------------------------------------------------------------------------------------------------------#
#Otra forma de hacerlo.
RutasListParadasTM['Rutas'] = RutasListParadasTM['Rutas'].str.rstrip(alphabet+alphabet.lower())
#------------------------------------------------------------------------------------------------------------------#
#Quitar espacio del principio.
RutasListParadasTM['Rutas'] = pd.Series([route.lstrip(" ") for route in RutasListParadasTM['Rutas']])
RutasListParadasTM
```

```
# La mayoria más no todas las rutas están bien tratadas, posteriormente se hará una limpeza mejor, por ahora esto nos sirve.

RutasListParadasTM.to_json("RutasListParadasTM.json",orient="records")
```

Se observa que hay que hacer el tratamiento posteriormente a la agrupación ya que se agruparon de manera incorrecta la primera vez porque se tomaron las estaciones tal y como estaban en el original y además también las rutas tal y como estaban.

Cabe resaltar que esto se pudo haber hecho desde un principio pero no se vio claramente al principio por eso se hizo posteriormente.

```
#Se hace el tratamiento de las rutas esta vez al dataSet Completo.

#Se hace un list Comprehension a la lista de Rutas, se eliminan los numeros y corchetes que aparecen al principio

DataSetTM['Rutas'] = pd.Series([route.lstrip("[0123456789]") for route in DataSetTM['Rutas']]) 
#lstrip Elimina los caracteres que se encuentran a la izquierda y que coinciden con el input.
#------------------------------------------------------------------------------------------------------------------#
#Otra forma de hacerlo.
DataSetTM['Rutas'] = DataSetTM['Rutas'].str.lstrip("[0123456789]")
```

```
#Se hace algo parecido para eliminar el punto de partida de la ruta que se encuentra al final del String. Se eliminan también espacios.
alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ._-éÁ "
DataSetTM['Rutas'] = pd.Series([route.rstrip(alphabet+alphabet.lower()) for route in DataSetTM['Rutas']])
#lstrip Elimina los caracteres que se encuentran a la derecha y que coinciden con el input.
#------------------------------------------------------------------------------------------------------------------#
#Otra forma de hacerlo.
DataSetTM['Rutas'] = DataSetTM['Rutas'].str.rstrip(alphabet+alphabet.lower())
#------------------------------------------------------------------------------------------------------------------#
#Quitar espacio del principio.
DataSetTM['Rutas'] = pd.Series([route.lstrip(" ") for route in DataSetTM['Rutas']])
DataSetTM
```
